#### Ejercicio 01. Análisis de datos básico con estructuras nativas de python.
Crea una lista con las calificaciones de 5 estudiantes: [8.5, 9.2, 7.8, 8.9, 10].

- Calcula el promedio de las calificaciones.

```python
# Importamos la librería NumPy
import numpy as np

# Asignamos la lista con la calificación de los 5 estudiantes a la variable "scores"
scores = [8.5, 9.2, 7.8, 8.9, 10]


# Calculamos el promedio de la lista de diferentes formas

# 1. Forma "python puro":
manual_average = sum(scores) / len(scores)


# 2. Media aritmética con NumPy (cuando todos los valores tienen el mismo peso):
arithmetic_mean = np.mean(scores)


# 3. Media ponderada con "NumPy" (cuando la nota final tiene el doble de peso para la media final):
# Tendríamos que añadir una nueva variable con los pesos de cada nota, por ejemplo:
weights = [1, 1, 1, 1, 2]

# Entonces ponemos como primer parámetro la lista de notas "scores" y como parámetro adicional "weights=weights"
weighted_average = np.average(scores, weights=weights)


# Mostramos en la consola los 3 cálculos
print(f"Media aritmética sin NumPy: {manual_average:.2f}")

print(f"Media aritmética: {arithmetic_mean:.2f}")

print(f"Media ponderada: {weighted_average:.2f}")
```

- Encuentra la calificación más alta y la más baja.

```python
# Para la calificación más alta utilizamos la función integrada max()
print(f"La calificación más alta es: {max(scores)}")

# Para la calificación más baja utilizamos la función integrada min()
print(f"La calificación más baja es: {min(scores)}")
```

#### Ejercicio 01. ¿Cuál es la casa más cara en todo el dataset? (★☆☆)

Imprime la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle del Prado, Nº20 es la más cara y su precio es de 5000000 USD`

```python
# Asignamos el precio máximo de la columna "df["price"]" con el método ".max()"
max_price = df["price"].max()

# Sacamos el index de la vivienda con el método ."idxmax()" y la asignamos a "max_price_index"
max_price_index = df['price'].idxmax()

# Asignamos a "address" la columna "address" de la línea que tenemos sacada anteriormente
address = df.loc[max_price_index, "address"]

# Mostramos en consola la f-string con sus variables correspondientes
print(f"La casa con dirección en la calle de {address} es la más cara y su precio es de {max_price} USD")
```

#### Ejercicio 02. ¿Cuál es la casa más barata del dataset? (★☆☆)

Este ejercicio es similar al ejercicio anterior, solo que ahora buscamos la casa con el precio más bajo. Recuerda imprimir la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle Alcalá, Nº58 es la más barata y su precio es de 12000 USD`

```python
# Primero buscamos el precio mínimo en la lista y lo asignamos a 'min_price'.
# min_price = df["price"].min()  -->  Problema: min_price = 0


# Como vemos que hay precios que son 0, los vamos a filtrar porque damos por sentado que no son datos que podamos tener en cuenta por el contexto.
# Para ello, creamos un nuevo DataFrame que nos filtre los 0 y lo asignamos a 'df_filtered_price'.
df_filtered_price = df[df['price'] > 0]


# Ahora que ya tenemos un dataframe que no tiene en cuenta las casas con un 'price' = 0, buscamos el precio mínimo y lo asignamos a 'min_price_filtered'.
min_price_filtered = df_filtered_price["price"].idxmin()


# Ya con el row correcto, podemos localizar el contenido de la celda 'datos_completos' con el método '.loc()'. 
datos_completos_min = df.loc[min_price_filtered]


# Ya tenemos la información para el ejercicio asi que mostramos en la terminal el resultado con la f-string correspondiente.
print(f"La casa con dirección en {datos_completos_min['address']} es la más barata y su precio es de {datos_completos_min['price']} USD")


# Utilizamos el SEPARATOR
print(SEPARATOR)


# Como extra, podemos verificar que el resultado es el esperado buscando otras lineas entre 1 y 600 incluido.
print("Verificación de lineas iguales o menores excluyendo 'df['price'] == 0' ")
df.loc[(df['price'] > 0) & (df['price'] <= 600)]
```

#### Ejercicio 03. ¿Cuál es la casa más grande y la más pequeña del dataset? (★☆☆)

Imprime la dirección y el área de las casas seleccionadas. Por ejemplo:

`La casa más grande está ubicada en Calle Gran Vía, Nº38 y su superficie es de 5000 metros`

`La casa más pequeña está ubicada en Calle Mayor, Nº12 y su superficie es de 200 metros`

Este ejercicio es similar al anterior, pero buscamos las casas más grandes y pequeñas en base a su superficie.

```python
# Filtramos las filas con "surface" mayor a 0
df_filtered_surface = df[df['surface'] > 0]

# Obtenemos el índice de la fila con la superficie más grande
surface_filtered_idx_max = df_filtered_surface["surface"].idxmax()

# Obtenemos el índice de la fila con la superficie más pequeña
surface_filtered_idx_min = df_filtered_surface["surface"].idxmin()


# Mostramos en consola los f-strings con "address" y "surface" de la casa más grande
print(f"La casa más grande está ubicada en {df.loc[surface_filtered_idx_max, 'address']} y su superficie es de {df.loc[surface_filtered_idx_max, 'surface']} metros")

# Mostramos en consola los f-strings con "address" y "surface" de la casa más pequeña
print(f"La casa más pequeña está ubicada en {df.loc[surface_filtered_idx_min, 'address']} y su superficie es de {df.loc[surface_filtered_idx_min, 'surface']} metros")
```

#### Ejercicio 06. Elimina los valores nulos (NAs) del dataset, si aplica (★★☆)

```python
# Eliminamos solo las columnas completamente vacías. Con "axis=1" trabajamos sobre columnas. Con "how='all'" eliminamos solo si las columnas están completamente vacías
df_clean_cols = df.dropna(axis=1, how='all')

# O también podemos eliminar filas donde faltan columnas importantes con "subset=['<columnas>,<a>,<eliminar>']"
df_clean_rows = df.dropna(subset=['price', 'address', 'surface'])

# Mostramos en pantalla el antes y el después
print(f"Original: {len(df)} filas")
print(SEPARATOR)
print(f"Sin columnas vacías: {len(df_clean_cols)} filas")
print(f"Sin filas incompletas: {len(df_clean_rows)} filas")
```

#### Ejercicio 07. ¿Cuál es la media de precios en la población de "Arroyomolinos (Madrid)"? (★★☆)

Imprima el valor obtenido de la columna level5.

```python
# Filtramos el DataFrame para quedarnos solo con las filas de Arroyomolinos
arroyomolinos = df[df['level5'] == 'Arroyomolinos (Madrid)']

# Calculamos la media de precios y la convertimos a entero para quitar decimales
am_price_mean = int(arroyomolinos['price'].mean())

# Mostramos el f-string en la terminal
print(f"La media de precios en Arroyomolinos (Madrid) es de {am_price_mean} USD")
```

#### Ejercicio 08. Trazar el histograma de los precios para la población de "Arroyomolinos (Madrid)" y explica qué observas (★★☆)

#### Ejercicio 01. Análisis de datos básico con estructuras nativas de python.
Crea una lista con las calificaciones de 5 estudiantes: [8.5, 9.2, 7.8, 8.9, 10].

- Calcula el promedio de las calificaciones.

```python
# Importamos la librería NumPy
import numpy as np

# Asignamos la lista con la calificación de los 5 estudiantes a la variable "scores"
scores = [8.5, 9.2, 7.8, 8.9, 10]


# Calculamos el promedio de la lista de diferentes formas

# 1. Forma "python puro":
manual_average = sum(scores) / len(scores)


# 2. Media aritmética con NumPy (cuando todos los valores tienen el mismo peso):
arithmetic_mean = np.mean(scores)


# 3. Media ponderada con "NumPy" (cuando la nota final tiene el doble de peso para la media final):
# Tendríamos que añadir una nueva variable con los pesos de cada nota, por ejemplo:
weights = [1, 1, 1, 1, 2]

# Entonces ponemos como primer parámetro la lista de notas "scores" y como parámetro adicional "weights=weights"
weighted_average = np.average(scores, weights=weights)


# Mostramos en la consola los 3 cálculos
print(f"Media aritmética sin NumPy: {manual_average:.2f}")

print(f"Media aritmética: {arithmetic_mean:.2f}")

print(f"Media ponderada: {weighted_average:.2f}")
```

#### Ejercicio 02. ¿Cuál es la casa más barata del dataset? (★☆☆)

Este ejercicio es similar al ejercicio anterior, solo que ahora buscamos la casa con el precio más bajo. Recuerda imprimir la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle Alcalá, Nº58 es la más barata y su precio es de 12000 USD`

```python
# Primero buscamos el precio mínimo en la lista y lo asignamos a 'min_price'.
min_price = min(list(df["price"]))
# Problema: min_price = 0


# Como vemos que hay precios que son 0, los vamos a filtrar porque damos por sentado que no son datos que podamos tener en cuenta por el contexto.
# Para ello, creamos un nuevo DataFrame que nos filtre los 0 y lo asignamos a 'df_filtered_price'.
df_filtered_price = df[df['price'] > 0]


# Ahora que ya tenemos un dataframe que no tiene en cuenta las casas con un 'price' = 0, buscamos el precio mínimo y lo asignamos a 'min_price_filtered'.
min_price_filtered = df_filtered_price["price"].idxmin()




# Ya con el row correcto, podemos localizar el contenido de la celda 'datos_completos' con el método '.loc()'. 
datos_completos_min = df.loc[min_price_filtered]


# Ya tenemos la información para el ejercicio asi que mostramos en la terminal el resultado con la f-string correspondiente.
print(f"La casa con dirección en {datos_completos_min['address']} es la más barata y su precio es de {datos_completos_min['price']} USD")
```
